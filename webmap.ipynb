# VOLCANOS

```python
# Importing the nessassery libraries
import folium
import pandas as pd
```

```python
### Reading info from Volcano data file
df = pd.read_csv('Volcanoes.txt')
df
```

### Getting information from dataframe and zip them

```python
### Getting information from dataframe and zip them
lat = list(df['LAT'])
lon = list(df['LON'])
name = list(df['NAME'])
elev = list(df['ELEV'])
coordinats = zip(lat, lon, name, elev)
elev[0]
```

```python
# Making a object of the folium Map
map = folium.Map(location=[45.305561, -75.932418],zoom_start=3)
```

```python
fg = folium.FeatureGroup(name= "My Map")
# fg.add_child(folium.Marker(location=[45.305561, -75.932418], popup='Hi Marker', icon=folium.Icon(color='red')))
# fg.add_child(folium.Marker(location=[65.305561, -51.932418], popup='Hi Marker', icon=folium.Icon(color='blue')))
```

```python
### Making map colorful by defining the color for each elevation, and call it in forLoop as icon colors
def colors(elev):
    color = ''
    if elev <= 1000:
        color = 'red'
    elif elev <= 2000:
        color = 'blue'
    elif elev <= 3000:
        color = 'yellow'
    else:
        color = 'green'
    return color
```

```python
for lat, lon, name, elev in coordinats:
    fg.add_child(folium.Marker(location=[lat, lon], popup= str(elev)+' , '+name, icon=folium.Icon(color= colors(elev))))
```

```python
map.add_child(fg)
```

```python
#### Saving the map to HTML file
map.save("Map1.html")
```
